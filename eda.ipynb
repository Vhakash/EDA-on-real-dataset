This code cell loads the various CSV datasets into pandas DataFrames. Each DataFrame corresponds to a different aspect of the e-commerce data, such as customer information, order details, products, sellers, payments, reviews, and product category translations.

```python
import pandas as pd

# Define the path to the dataset directory
path = '/content/daata'

# Load the datasets
customers_df = pd.read_csv(f'{path}/olist_customers_dataset.csv')
orders_df = pd.read_csv(f'{path}/olist_orders_dataset.csv')
order_items_df = pd.read_csv(f'{path}/olist_order_items_dataset.csv')
products_df = pd.read_csv(f'{path}/olist_products_dataset.csv')
sellers_df = pd.read_csv(f'{path}/olist_sellers_dataset.csv')
order_payments_df = pd.read_csv(f'{path}/olist_order_payments_dataset.csv')
order_reviews_df = pd.read_csv(f'{path}/olist_order_reviews_dataset.csv')
product_category_translation_df = pd.read_csv(f'{path}/product_category_name_translation.csv')

print("All datasets loaded successfully.")
```

MERGING RELATED COLUMS

This cell merges the `orders_df` and `customers_df` DataFrames on the `customer_id` column. It then displays the first 5 rows of the resulting merged DataFrame.

```python
# Merge orders and customers dataframes
orders_customers_df = pd.merge(orders_df, customers_df, on='customer_id', how='left')

print("\nMerged orders_df and customers_df:")
display(orders_customers_df.head())
```

This cell merges the `order_items_df` and `products_df` DataFrames on the `product_id` column. It then displays the first 5 rows of the resulting merged DataFrame.

```python
# Merge order_items and products dataframes
order_items_products_df = pd.merge(order_items_df, products_df, on='product_id', how='left')

print("\nMerged order_items_df and products_df:")
display(order_items_products_df.head())
```

This cell merges the `order_items_products_df` (which contains order item and product information) and `sellers_df` DataFrames on the `seller_id` column. It then displays the first 5 rows of the resulting merged DataFrame.

```python
# Merge order_items_products and sellers dataframes
order_items_products_sellers_df = pd.merge(order_items_products_df, sellers_df, on='seller_id', how='left')

print("\nMerged order_items_products_df and sellers_df:")
display(order_items_products_sellers_df.head())
```

This cell merges the `orders_customers_df` (which contains order and customer information) and `order_payments_df` DataFrames on the `order_id` column. It then displays the first 5 rows of the resulting merged DataFrame.

```python
# Merge orders_customers and order_payments dataframes
orders_customers_payments_df = pd.merge(orders_customers_df, order_payments_df, on='order_id', how='left')

print("\nMerged orders_customers_df and order_payments_df:")
display(orders_customers_payments_df.head())
```

This cell merges the `orders_customers_df` (which contains order and customer information) and `order_reviews_df` DataFrames on the `order_id` column. It then displays the first 5 rows of the resulting merged DataFrame.

```python
# Merge orders_customers and order_reviews dataframes
orders_customers_reviews_df = pd.merge(orders_customers_df, order_reviews_df, on='order_id', how='left')

print("\nMerged orders_customers_df and order_reviews_df:")
display(orders_customers_reviews_df.head())
```

This cell merges the `order_items_products_df` (which contains order item and product information) and `product_category_translation_df` DataFrames on the `product_category_name` column. It then displays the first 5 rows of the resulting merged DataFrame.

```python
# Merge order_items_products and product_category_translation dataframes
order_items_products_translated_df = pd.merge(order_items_products_df, product_category_translation_df, on='product_category_name', how='left')

print("\nMerged order_items_products_df and product_category_translation_df:")
display(order_items_products_translated_df.head())
```
